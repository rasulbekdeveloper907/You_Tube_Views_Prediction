```python
df.drop('Unnamed: 0.1', axis=1,inplace=True)
```

```python
df.drop('Unnamed: 0', axis=1,inplace=True)
```

```python
df.info()
```

```python
target = 'CategoryName'
X = df.drop(['CategoryName', target], axis=1)
y = df[target]
```

```python
# 1Ô∏è‚É£ Target ustun (y)
y = df['CategoryName']

# 2Ô∏è‚É£ X (features)
X = df.drop(columns=['CategoryName'])

# 3Ô∏è‚É£ Kategorik ustunlar (katta ehtimol bilan bu quyidagilar)
cat_cols = [
    'Country', 
    'Region', 
    'CategoryID', 
    'Cluster', 
    'CategoryID_enc',
    'VideoCluster', 
    'Channel_Cluster', 
    'Video_Cluster',
    'Country_Cluster_x', 
    'Category_Cluster_x'
]

# 4Ô∏è‚É£ Sonli ustunlar ‚Äî yuqoridagilardan tashqari hammasi
num_cols = [col for col in X.columns if col not in cat_cols]

print("üîπ Kategorik ustunlar:")
print(cat_cols)
print("\nüîπ Sonli ustunlar:")
print(num_cols)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols)
    ],
    remainder='passthrough'  # numeric columns remain unchanged
)
```

```python
preprocessor
```

```python
pipeline = Pipeline([
    ('preprocess', preprocessor),
    ('regressor', RandomForestClassifier(random_state=42))
])
```

```python
pipeline
```

```python
pipeline.fit(X, y)
```

# Qoshimcha ishlar

```python
import joblib

# Pipeline'ni saqlash
joblib.dump(
    pipeline,
    r"C:\Users\Rasulbek907\Desktop\Project_MP\Models\Simple_Models\RandomForestClassifier.joblib"
)

print("‚úÖ Model muvaffaqiyatli saqlandi!")
```
