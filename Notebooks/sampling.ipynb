```python
import numpy as np
import pandas as pd
from collections import Counter
from imblearn.over_sampling import RandomOverSampler, SMOTE, ADASYN
```

```python
x = df.drop(columns=['CategoryName'])  
y = df['CategoryName']             
```

```python
x
```

```python
y
```

```python
print("Original class distribution:", Counter(y))
```

# Random Oversampling

```python
ros = RandomOverSampler(random_state=42)
X_ros, y_ros = ros.fit_resample(x, y)
```

```python
print("\nRandom Oversampling class distribution:", Counter(y_ros))
print("Random Oversampling samples:\n", X_ros.values.flatten())
```

```python
X_ros, y_ros
```

# SMOTE

```python
smote = SMOTE(random_state=42)
X_smote, y_smote = smote.fit_resample(x, y)
```

```python
print("\nSMOTE class distribution:", Counter(y_smote))
print("SMOTE samples:\n", X_smote.values.flatten())
```

```python
X_smote, y_smote
```

#  ADASYN

```python
adasyn = ADASYN(sampling_strategy='minority', random_state=42)
X_ada, y_ada = adasyn.fit_resample(x, y)
```

```python
print("\nADASYN class distribution:", Counter(y_ada))
print("ADASYN samples:\n", X_ada.values.flatten())
```
